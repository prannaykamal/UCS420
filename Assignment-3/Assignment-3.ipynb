```python
#Q6
dd={
    'Employee_ID':[101,102,103,104,105],
    'Name':['Alice','Bob','Charlie','Diana','Edward'],
    'Department':['HR','IT','IT','Marketing','Sales'],
    'Age':[29,34,41,28,38],
    'Salary':[50000,70000,65000,55000,60000],
    'Years_of_Experience':[4,8,10,3,12],
    'Joining_Date':['2020-03-15','2017-07-19','2013-06-01','2021-02-10','2010-11-25'],
    'Gender':['Female','Male','Male','Female','Male'],
    'Bonus':[5000,7000,6000,4500,5000],
    'Rating':[4.5,4.0,3.8,4.7,3.5]
}
dataset=pd.DataFrame(dd)
dataset.to_csv('employees.csv')
print(dataset)
```

```python
print("average salary is ",dataset['Salary'].mean())
print("total bonus is ",dataset['Bonus'].sum())
print("age of youngest employee is ",dataset['Age'].min())
print("highest rating is ",dataset['Rating'].max())
```

```python
sorteddataset=dataset.sort_values(by='Salary',ascending=False)
print(sorteddataset)
```

```python
def categorize_rating(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'
dataset['Performance_Category']=dataset['Rating'].apply(categorize_rating)
print(dataset)
```

```python
missingval=dataset.isnull().sum()
print(missingval)
```

```python
dataset.rename(columns={'Employee_ID':'ID'}, inplace=True)
print(dataset)
```

```python
print(dataset[(dataset['Years_of_Experience']>5)])
print(dataset[(dataset['Department']=='IT')])
```
