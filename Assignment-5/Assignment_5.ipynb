```python
#Q1
import numpy as np
gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')
print("sum of all elements",np.sum(gfg))
print("row-wise sum",np.sum(gfg,axis=1))
print("coloumn-wise sum",np.sum(gfg,axis=0))
```

```python
#Q2(a)
array=np.array([10,52,62,16,16,54,453])
s=np.sort(array)
print("sorted array",s)
print("indices of sorted array",np.argsort(array))
print("smallest 4:",s[:4])
print("largest 5:",s[-5:])
```

```python
#Q2(b)
array=np.array([1.0,1.2,2.2,2.0,3.0,2.0])
integers=array[array==array.astype(int)]
floats=array[array!=array.astype(int)]
print("integer elements",integers)
print("float elements",floats)
```

```python
#Q3
X=ord('P')+ord('K')
sales=np.array([X,X+50,X+100,X+150,X+200])
print("sales",sales)
rate=((X%5)+5)/100
print("tax rate",rate)
taxed_sales=sales*(1+rate)
print("sales after applying tax",taxed_sales)
discounted_sales=np.where(sales<X+100,sales*0.95,sales*0.90)
print("sales after appling discount",discounted_sales)
matrix=np.resize(sales,(3,5))
print("matrix\n",matrix)
rates=np.array([1.02,1.0404,1.061208])
for i in range(3):
  matrix[i]=matrix[i]*rates[i]
print("matrix after increasing sales by 2% per week\n",matrix)
```

```python
#Q4
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)
y1=x**2
y2=np.sin(x)
y3=np.exp(x)
y4=np.log(np.abs(x)+1)
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.plot(x,y1,label='y=x^2')
plt.title('y=x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.subplot(2,2,2)
plt.plot(x,y2,label='y=sin(x)')
plt.title('y=sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.subplot(2,2,3)
plt.plot(x,y3,label='y=e^x')
plt.title('y=e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.subplot(2,2,4)
plt.plot(x,y4,label='y=log(|x|+1)')
plt.title('y=log(|x|+1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.show()
```
