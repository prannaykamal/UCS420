```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Q1
M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, 'r--', label=r'y = Mx^2')
plt.plot(x, y2, 'b-', label=r'y = Msin(x)')
plt.legend()
plt.grid()
plt.title('Mathematical Functions Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

```python
#Q2
data={
    'Subject':['English','Maths','Physics','Chemistry','Biology'],
    'Score':[85,92,90,81,78]
}
df=pd.DataFrame(data)
print(df)
colors=sns.color_palette('viridis',len(df))
plt.figure(figsize=(10,6))
bar_plot=sns.barplot(x='Subject',y='Score',data=df,palette=colors)
for index,row in df.iterrows():
    bar_plot.text(index,row.Score+1,round(row.Score,2),color='black',ha="center")
plt.title('Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
#Q3
rno=102317125
np.random.seed(rno)
data=np.random.randn(50)

fig,axes=plt.subplots(2,2,figsize=(10,8))
axes=axes.flatten()

axes[0].plot(np.cumsum(data),'g-',label='Cumulative Sum')
axes[0].set_title('Cumulative Sum Plot')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Cumulative Sum')
axes[0].legend()
axes[0].grid()

axes[1].scatter(range(50),data,c='red',marker='o',alpha=0.6)
axes[1].set_title('Scatter Plot with Noise')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Random Values')
axes[1].grid()

plt.show()
```

```python
# Q4

data=pd.read_csv("company_sales_data.csv")

#1
plt.figure(figsize=(10,5))
plt.plot(data['month_number'],data['total_profit'],'bo-',label='Total Profit')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.legend()
plt.grid()
plt.show()

#2
data.plot(x='month_number',y=data.columns[1:-1],kind='line',figsize=(10,5))
plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(title='Products')
plt.grid()
plt.show()

#3
data.iloc[:,1:].sum().plot(kind='bar',figsize=(10,5),color='blue')
plt.title('Total Sales by Feature')
plt.xlabel('Features')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--')
plt.show()
```
